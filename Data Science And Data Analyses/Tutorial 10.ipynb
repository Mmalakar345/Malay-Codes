# axis label & heading

```python
plt.plot([1,2,3],[8,9,7])
plt.title("Malay's Data")
plt.ylabel("This is Y axis")
plt.xlabel("This is X axis")
plt.show()
```

# Style

```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
x = [1,5,9]
y = [23,55,67]
x2 = [11,22,13]
y2 = [12,33,44]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

# Legend tag

```python
style.use('ggplot')
x = [1,5,9]
y = [23,55,67]
x2 = [11,22,13]
y2 = [12,33,44]
plt.plot(x,y,label='first')
plt.plot(x2,y2,label='second')
plt.legend()
plt.show()
```

# line depth

```python
style.use('ggplot')
x = [1,5,9]
y = [23,55,67]
x2 = [11,22,13]
y2 = [12,33,44]
plt.plot(x,y,label='first',linewidth=5)
plt.plot(x2,y2,label='second',linewidth=3)
plt.legend()
plt.show()
```

# Scatter

```python
style.use('ggplot')
x = [1,5,9]
y = [23,55,67]
x2 = [11,22,13]
y2 = [12,33,44]
plt.scatter(x,y,label='first',linewidth=5)
plt.plot(x2,y2,label='second',linewidth=3)
plt.legend()
plt.show()
```

# histogram

```python
style.use('ggplot')
x = [2,5,9]
y = [23,55,67]
plt.hist(x,y,label='first',linewidth=5)
plt.legend()
plt.show()
```
